# **3 Image Compression and Decompression using PCA.**

## **1. Load and Prepare Data:**
**• Fetch an image of you choice.{If colour convert to grayscale}**<br>
**• Center the dataset- Standaridze the Data.**<br>
**• Calculate the covaraince matrix of the Standaridze data.**

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# 1. Load and prepare data
img_path = "/content/drive/My Drive/AI and Machine Learning/Workshop 1/cameraman.png"
img = Image.open(img_path).convert("L")        # ensure grayscale

# Convert to NumPy array (float for PCA)
X = np.array(img, dtype=float)                 # shape (H, W)

# Show original grayscale image
plt.imshow(X, cmap="gray")
plt.title("Original Grayscale Cameraman")
plt.axis("off")
plt.show()

# 2. Center / standardize the data
# Treat each column (pixel column) as a feature
mean_cols = np.mean(X, axis=0)                 # mean of each column
std_cols  = np.std(X, axis=0) + 1e-8           # std of each column (avoid divide-by-zero)

X_std = (X - mean_cols) / std_cols             # standardized data, mean≈0, var≈1 per column

# 3. Covariance matrix of standardized data
# columns are variables => rowvar=False
cov_mat = np.cov(X_std, rowvar=False)

print("Image shape:", X.shape)
print("Standardized data shape:", X_std.shape)
print("Covariance matrix shape:", cov_mat.shape)
```

## **2. Eigen Decomposition and Identifying Principal Components:**
**• Compute Eigen Values and Eigen Vectors.**<br>
**• Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to the highest eigenvalues.**<br>
**• Identify the Principal Components with the help of cumulative Sum plot.**

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Eigen decomposition of covariance matrix
# cov_mat is (W, W) from previous code
eigen_vals, eigen_vecs = np.linalg.eigh(cov_mat)   # symmetric matrix => eigh is stable

# 2. Sort eigenvalues (and eigenvectors) in descending order
idx = np.argsort(eigen_vals)[::-1]      # indices for sorting high → low
eigen_vals = eigen_vals[idx]
eigen_vecs = eigen_vecs[:, idx]

print("Top 5 eigenvalues:", eigen_vals[:5])
```

```python
# 3. Explained variance ratio for each component
total_var = np.sum(eigen_vals)
explained_var_ratio = eigen_vals / total_var

# 4. Cumulative explained variance
cum_explained = np.cumsum(explained_var_ratio)

# 5. Plot cumulative sum to identify k
plt.figure(figsize=(6,4))
plt.plot(range(1, len(cum_explained)+1), cum_explained, marker="o")
plt.xlabel("Number of principal components (k)")
plt.ylabel("Cumulative explained variance")
plt.title("Cumulative Explained Variance vs k")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```

## **3. Reconstruction and Experiment:**
**• Reconstruction: Transform the original data by multiplying it with the selected eigenvectors(PCs) to obtain a lower-dimensional representation.**<br>
**• Experiments: Pick Four different combination of principal components with various explained variance value and compare the result.**<br>
**• Display the Results and Evaluate.**

```python
import numpy as np

def pca_reconstruct(X_std, mean_cols, std_cols, eigen_vecs, k):
    # Select first k eigenvectors (principal components)
    W_k = eigen_vecs[:, :k]              # shape (W, k)

    # Projection to k-dimensional space (scores)
    Z = X_std @ W_k                      # shape (H, k)

    # Back-project to original feature space
    X_std_recon = Z @ W_k.T              # shape (H, W)

    # De-standardize: undo standardization
    X_recon = X_std_recon * std_cols + mean_cols

    # Clip to valid image range and convert to uint8
    X_recon = np.clip(X_recon, 0, 255).astype(np.uint8)
    return X_recon
```

```python
import matplotlib.pyplot as plt

k_values = [10, 20, 50, 100]   # replace by your own choices

recons = []
for k in k_values:
    X_k = pca_reconstruct(X_std, mean_cols, std_cols, eigen_vecs, k)
    recons.append(X_k)
```
