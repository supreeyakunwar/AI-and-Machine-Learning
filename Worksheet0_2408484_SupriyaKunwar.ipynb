## Task- 1:

**Create a Python program that converts between different units of measurement.**<br>**• The program should**:<br>
**1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).**<br>
**2. Ask the user to input the value to be converted.**<br>
**3. Perform the conversion and display the result.<br>**
**4. Handle potential errors, such as invalid input or unsupported conversion types.**

```python
# Task-1: Unit Converter

# Requirement 1: Function with docstring
def convert_units(conversion_type, value, direction):
    """
    Convert between length, weight, or volume.

    Parameters
    ----------
    conversion_type : str
        "length", "weight", or "volume".
    value : float
        The numeric value to be converted.
    direction : str
        For length: "m_to_ft" or "ft_to_m"
        For weight: "kg_to_lbs" or "lbs_to_kg"
        For volume: "L_to_gal" or "gal_to_L"

    Returns
    -------
    float
        Converted value.
    """
    if conversion_type == "length":
        if direction == "m_to_ft":
            return value * 3.28084               # meters -> feet
        elif direction == "ft_to_m":
            return value / 3.28084               # feet -> meters

    elif conversion_type == "weight":
        if direction == "kg_to_lbs":
            return value * 2.20462               # kg -> lbs
        elif direction == "lbs_to_kg":
            return value / 2.20462               # lbs -> kg

    elif conversion_type == "volume":
        if direction == "L_to_gal":
            return value * 0.264172              # L -> gal
        elif direction == "gal_to_L":
            return value / 0.264172              # gal -> L

    # Requirement 4: unsupported conversion type/direction
    raise ValueError("Unsupported conversion type or direction.")

# MAIN PROGRAM


try:
    # Task point 1 + Requirement 3:
    # Prompt the user to choose the type of conversion
    print("Choose the type of conversion:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")
    conversion_choice = input("Enter 1, 2, or 3: ")

    if conversion_choice == "1":
        conversion_type = "length"
        direction = input(
            "Enter conversion ('m_to_ft' for meters to feet, "
            "'ft_to_m' for feet to meters): "
        )

    elif conversion_choice == "2":
        conversion_type = "weight"
        direction = input(
            "Enter conversion ('kg_to_lbs' for kilograms to pounds, "
            "'lbs_to_kg' for pounds to kilograms): "
        )

    elif conversion_choice == "3":
        conversion_type = "volume"
        direction = input(
            "Enter conversion ('L_to_gal' for liters to gallons, "
            "'gal_to_L' for gallons to liters): "
        )

    else:
        # invalid menu choice
        raise ValueError("Invalid conversion type choice.")

    # Task point 2:
    # Ask the user to input the value to be converted
    value = float(input("Enter the value to be converted: "))

    # Task point 3:
    # Perform the conversion and display the result
    result = convert_units(conversion_type, value, direction)
    print(f"Converted value: {result:.4f}")

# Task point 4 + Requirement 2:
# Handle invalid input or unsupported conversion types
except ValueError as e:
    print("Error:", e)
```

## Task- 2:


**Create a Python program that performs various mathematical operations on a list of numbers.**<br>
**• The Program should:**<br>
**1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum
of the numbers).<br>**
**2. Ask the user to input a list of numbers (separated by spaces).<br>**
**3. Perform the selected operation and display the result.**<br>
**4. Handle potential errors, such as invalid input or empty lists.**

```python
# Task-2: List Operations

# Requirement 1 + 4: functions with docstrings

def calc_sum(numbers):
    """Return the sum of a list of numbers."""
    return sum(numbers)


def calc_average(numbers):
    """Return the average of a list of numbers. Raises ValueError if list is empty."""
    if not numbers:
        raise ValueError("Cannot compute average of an empty list.")
    return sum(numbers) / len(numbers)


def calc_max(numbers):
    """Return the maximum value in a list of numbers. Raises ValueError if list is empty."""
    if not numbers:
        raise ValueError("Cannot compute maximum of an empty list.")
    return max(numbers)


def calc_min(numbers):
    """Return the minimum value in a list of numbers. Raises ValueError if list is empty."""
    if not numbers:
        raise ValueError("Cannot compute minimum of an empty list.")
    return min(numbers)

# MAIN PROGRAM


try:
    # 1. Prompt the user to choose an operation
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")
    op_choice = input("Enter 1, 2, 3, or 4: ")

    # 2. Ask the user to input a list of numbers (separated by spaces)
    raw = input("Enter numbers separated by spaces: ")
    numbers = list(map(float, raw.split()))   # convert input to list of floats[web:62][web:68]

    if not numbers:
        raise ValueError("The list of numbers is empty.")

    # 3. Perform the selected operation and display the result
    if op_choice == "1":
        result = calc_sum(numbers)
        print("Sum:", result)
    elif op_choice == "2":
        result = calc_average(numbers)
        print("Average:", result)
    elif op_choice == "3":
        result = calc_max(numbers)
        print("Maximum:", result)
    elif op_choice == "4":
        result = calc_min(numbers)
        print("Minimum:", result)
    else:
        raise ValueError("Invalid operation choice.")

# 4. Handle potential errors (invalid input or empty list)
except ValueError as e:
    print("Error:", e)
```

**1. Flatten a Nested List:**<br>
**Write a Python function that takes a nested list and flattens it into a single list, where all the elements are in a single dimension.**<br>
**• Requirements:–**<br> 
**Define a function flatten(lst) that takes a nested list lst and returns a flattened version of the list.**<br>
**Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].**

```python
def flatten(lst):
    """
    Flatten a nested list with one level of nesting.

    Parameters
    ----------
    lst : list
        Nested list, e.g. [[1, 2], [3, 4], [5]].

    Returns
    -------
    list
        Single flat list of elements.
    """
    flat = []
    for sub in lst:
        flat.extend(sub)   # add all elements from each inner list
    return flat


# Example test
nums = [[1, 2], [3, 4], [5]]
print(flatten(nums))   
```

**3. Sum of All Elements in a Nested List:**<br>
**Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).**<br>
**• Requirements:–**<br>
**Define a function sum nested(lst) that takes a nested list lst and returns the sum of all the elements.**<br>
**Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 21.**

```python
def sum_nested(lst):
    """
    Return the sum of all numbers in a nested list, regardless of depth.

    Parameters
    ----------
    lst : list
        Nested list containing ints/floats or other lists.

    Returns
    -------
    int or float
        Sum of all numeric elements.
    """
    total = 0
    for element in lst:
        if isinstance(element, list):
            # if it's a list, sum its contents recursively
            total += sum_nested(element)
        else:
            # if it's a number, add directly
            total += element
    return total


# Example test (from the question)
nums = [[1, 2], [3, [4, 5]], 6]
print(sum_nested(nums))   

# Another example
nums2 = [1, [2, 3], [[4], 5]]
print(sum_nested(nums2))  
```

**4. Remove Specific Element from a Nested List:**<br>
**Write a Python function that removes all occurrences of a specific element from a nested list.**<br>
**• Requirements:–**<br> 
**Define a function remove element(lst, elem) that removes elem from lst and returns themodified list.**<br>
**Example: Fortheinput lst = [[1, 2], [3, 2], [4, 5]]andelem = 2,theoutputshould be [[1], [3], [4, 5]]**

```python
def remove_element(lst, elem):
    """
    Remove all occurrences of `elem` from a (possibly nested) list.

    Parameters
    ----------
    lst : list
        Nested list.
    elem : any
        Element to remove.

    Returns
    -------
    list
        New nested list with all occurrences of elem removed.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            # Clean the inner list recursively
            cleaned = remove_element(item, elem)
            result.append(cleaned)
        else:
            # Only keep items that are not equal to elem
            if item != elem:
                result.append(item)
    return result


# Example test (from the question)
lst = [[1, 2], [3, 2], [4, 5]]
elem = 2
print(remove_element(lst, elem))  

# Another example
lst2 = [1, [2, 3, 2], [4, [2, 5]]]
print(remove_element(lst2, 2))    
```

**5. Find the Maximum Element in a Nested List:**<br>
**Write a Python function that finds the maximum element in a nested list (regardless of depth).**<br>
**• Requirements:–**<br> 
**Define a function find max(lst) that takes a nested list lst and returns the maximum element.**<br>
**Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.**

```python
def find_max(lst):
    """
    Find the maximum numeric element in a nested list, regardless of depth.

    Parameters
    ----------
    lst : list
        Nested list containing numbers or other lists.

    Returns
    -------
    int or float
        Maximum value found in the nested structure.
    """
    current_max = None

    for element in lst:
        if isinstance(element, list):
            # Recursively find max in the sublist
            sub_max = find_max(element)
            if current_max is None or sub_max > current_max:
                current_max = sub_max
        else:
            # Element is a number
            if current_max is None or element > current_max:
                current_max = element

    return current_max


# Example test (from the question)
nums = [[1, 2], [3, [4, 5]], 6]
print(find_max(nums))   

# Another example
nums2 = [1, [10, 3], [[7, 20], 4]]
print(find_max(nums2)) 
```

**6. Count Occurrences of an Element in a Nested List:**<br>
**Write a Python function that counts how many times a specific element appears in a nested list.**<br>
**• Requirements:–**<br>
**Define a function count occurrences(lst, elem) that counts the occurrences of elem in the nested list lst.**<br>
**Example: Fortheinput lst = [[1, 2], [2, 3], [2, 4]]andelem = 2,the output should be 3.**

```python
def count_occurrences(lst, elem):
    """
    Count how many times `elem` appears in a nested list.

    Parameters
    ----------
    lst : list
        Nested list.
    elem : any
        Element to count.

    Returns
    -------
    int
        Number of occurrences of elem in the nested structure.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            # Recurse into inner list
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count


# Example test (from the question)
lst = [[1, 2], [2, 3], [2, 4]]
elem = 2
print(count_occurrences(lst, elem))   

# Another example
lst2 = [1, [2, 3, 2], [4, [2, 5]]]
print(count_occurrences(lst2, 2))   
```

**7. Flatten a List of Lists of Lists:**<br>
**Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.**<br>
**• Requirements:–**<br>
**Define a function deep flatten(lst) that takes a deeply nested list lst and returns a single flattened list.**<br>
**Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be [1, 2, 3, 4, 5, 6, 7, 8].**

```python
def deep_flatten(lst):
    """
    Flatten a deeply nested list into a single list.

    Parameters
    ----------
    lst : list
        List that may contain other lists at any depth.

    Returns
    -------
    list
        One-dimensional list with all elements.
    """
    flat = []
    for item in lst:
        if isinstance(item, list):
            flat.extend(deep_flatten(item))   # recurse into sublists
        else:
            flat.append(item)
    return flat


# Example test (from the question)
nums = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(deep_flatten(nums))   

# Another example
nums2 = [1, [2, [3, 4], 5], 6]
print(deep_flatten(nums2))  
```

**8. Nested List Average:**<br>
**Write a Python function that calculates the average of all elements in a nested list.**<br>
**• Requirements:–**<br> 
**Define a function average_nested(lst) that takes a nested list lst and returns the average of all the elements.**<br>
**Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.**

```python
def average_nested(lst):
    """
    Return the average of all numeric elements in a nested list.

    Parameters
    ----------
    lst : list
        Nested list containing numbers or other lists.

    Returns
    -------
    float
        Average of all numbers in the nested structure.
    """
    def helper(sublist):
        total = 0
        count = 0
        for item in sublist:
            if isinstance(item, list):
                sub_total, sub_count = helper(item)
                total += sub_total
                count += sub_count
            else:
                total += item
                count += 1
        return total, count

    total, count = helper(lst)
    if count == 0:
        raise ValueError("Cannot compute average of an empty nested list.")
    return total / count


# Example test (from the question)
nums = [[1, 2], [3, 4], [5, 6]]
print(average_nested(nums))   

# Another example
nums2 = [1, [2, 3], [[4], 5]]
print(average_nested(nums2))  
```

### Problem- 4: Matrix Operations:
###  For the following arrays:
### A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);

#### Prove following with Numpy:

**1. Prove A.A−1 = I.**

```python
import numpy as np

# Given matrices
A = np.array([[3, 4],
              [7, 8]])

# 1. Compute the inverse of A
A_inv = np.linalg.inv(A)   # uses NumPy's linear algebra module [web:326]

# 2. Multiply A and A_inv to check A · A^{-1} = I
I = A @ A_inv              # or: I = np.dot(A, A_inv)

print("A =\n", A)
print("\nA inverse =\n", A_inv)
print("\nA · A^{-1} =\n", I)
```

**2. Prove AB̸ = BA.**

```python
import numpy as np

A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

# Matrix products
AB = A @ B           # or np.dot(A, B)
BA = B @ A           # or np.dot(B, A)

print("A =\n", A)
print("B =\n", B)

print("\nA @ B =\n", AB)
print("\nB @ A =\n", BA)

print("\nAre A @ B and B @ A equal? ->", np.array_equal(AB, BA))
```

**3. Prove (AB)T = BTAT.**

```python
import numpy as np

A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

# Left-hand side: (AB)^T
AB = A @ B
lhs = AB.T                 # transpose of the product

# Right-hand side: B^T A^T
rhs = B.T @ A.T

print("A =\n", A)
print("B =\n", B)

print("\n(AB)^T =\n", lhs)
print("\nB^T A^T =\n", rhs)

print("\nAre (AB)^T and B^T A^T equal? ->", np.array_equal(lhs, rhs))
```
